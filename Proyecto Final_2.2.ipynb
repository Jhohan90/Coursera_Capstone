```python
# Importar Libreria
import pandas as pd
```

```python
# Importar DataFrame (df3) de Proyecto Final_2.1
df3=pd.read_excel('df3.xlsx')
df3
```

```python
# Eliminar columna innecesaria
df3.drop(['Unnamed: 0'], axis = 1, inplace=True)
df3
```

```python
# Explorar y agrupar los vecindarios de los municipios que contienen la palabra 'Toronto'
toronto_data = df3[df3.Borough.str.contains('Toronto')].reset_index(drop=True)
toronto_data
```

```python
# Importar libreria para convertir una dirección en valores de latitud y longitud
from geopy.geocoders import Nominatim 
```

```python
# Establecer coordenadas
address = 'Toronto'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# Importar librería para graficar mapas 
import folium 
```

```python
# crear un mapa de Toronto usando los valores de latitud y longitud
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# añadir los marcadores al mapa
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Borough']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
